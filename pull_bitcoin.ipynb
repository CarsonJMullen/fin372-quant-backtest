```python
import yfinance as yf
import pandas as pd

# Fetch historical data for BTC-USD
data = yf.download("BAC")

# Calculate daily returns
# Daily return = (Today's Close / Yesterday's Close) - 1
data['Daily Return'] = data['Close'].pct_change()

# Remove any rows with NaN values which might be present in the first row
data.dropna(subset=['Daily Return'], inplace=True)
```
