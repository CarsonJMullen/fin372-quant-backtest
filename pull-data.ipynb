```python
import yfinance as yf
import pandas as pd
```

```python
def rm_us(str):
    return str.split(" US")[0]

us_issuances = pd.read_csv('data/Bank US Secondaries.csv')
us_issuances['Issuer Ticker'] = us_issuances['Issuer Ticker'].apply(rm_us)
us_issuances = us_issuances[['Announced Date', 'Issuer Ticker', 'Offer Size (M)']].rename(columns = {'Announced Date': 'Date', 'Issuer Ticker': 'Ticker'})
us_issuances.to_csv('data/signal-raw.csv', index=False)

unique_tickers = list(us_issuances['Ticker'].unique())

# For some reason we are unable to append BTC and download in one file: run into yfinance error so have to download separately
data = yf.download(unique_tickers)
btc_data = yf.download("BTC-USD")
```

```python
yahoo_adjclose = data['Adj Close']
yahoo_adjclose.dropna(axis=1, how='all', inplace=True)
```

```python
yahoo_adjclose['BTC'] = btc_data['Adj Close']
```

```python
data['Adj Close'].pct_change()
```

```python
prices_df = pd.melt(data['Adj Close'].reset_index(), id_vars=['Date'], var_name='Ticker', value_name='Adj Close').dropna()
prices_df
```

```python
data.reset_index()
```

```python
btc_data = btc_data['Adj Close'].reset_index()
btc_data['Ticker'] = "BTC-USD"
btc_data
```

```python
prices_df = pd.concat([prices_df, btc_data]).reset_index(drop=True)
```

```python
prices_df
```

```python
prices_df['ret'] = prices_df.groupby('Ticker')['Adj Close'].pct_change()
```

```python
mkt_rf_df = pd.read_csv('data/mkt_rf_daily.csv')
date = pd.to_datetime(mkt_rf_df.loc[:,'date'],format='%Y%m%d')
```

```python
prices_df = prices_df[(prices_df['Date'].isin(date)) & (prices_df['Date'] >= '2014-09-17')]
prices_df
```

```python
# Save the DataFrame to CSV format
prices_df[prices_df['Ticker'] == "BTC-USD"].to_csv('data/bitcoin.csv', index=False)
prices_df.to_csv('data/yf-prices.csv', index=False)
```
