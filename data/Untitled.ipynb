```python
import pandas as pd

# Convert 'Announced Date' in bases_df and 'date' in wrd_df to datetime objects
bases_df['Announced Date'] = pd.to_datetime(bases_df['Announced Date'])
wrds_df['date'] = pd.to_datetime(wrds_df['date'])

bases_df = bases_df.sort_values('Announced Date')
wrds_df = wrds_df.sort_values('date')

# Perform an asof merge with a tolerance of 7 days
df = pd.merge_asof(bases_df, wrds_df[['date', 'TICKER', 'MKTCAP']],
                   left_on='Announced Date', right_on='date',
                   by='TICKER', tolerance=pd.Timedelta(days=7), direction='nearest')
```
